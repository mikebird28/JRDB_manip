```python
import sqlite3
import pandas as pd
from sklearn.decomposition import PCA
```

```python
import dataset2
import util
```

```python
db_path = "db/output_v3.db"
config = util.get_config("config/config.json")
x,y = dataset2.load_dataset(db_path,config.features,"win")
```

```python
train_x,test_x,train_y,test_y = dataset2.split_with_race(x,y)
```

```python
train_x.describe()
```

```python
train_x = dataset2.fillna_mean(train_x,"horse")
```

```python
pca_x,pca_y = dataset2.pad_race(train_x,train_y)
pca_x = dataset2.flatten_race(pca_x)
```

```python
pca_df = pd.DataFrame(pca_x)
```

```python
pca = PCA(n_components = 10)
```

```python
pca_idx = pca_x["info_race_id"]
pcafu_x,pcafu_y = dataset2.for_use(pca_df,pca_y)
pca.fit(pcafu_x)
```

```python
print(sum(pca.explained_variance_ratio_))
```

```python
pcaed_df = pd.DataFrame(pca.transform(pcafu_x))
pcaed_df = pd.concat([pd.DataFrame(pcaed_df),pca_idx],axis = 1)
```

```python
pcaed_df.describe()
```

```python
print(len(train_y) - train_y.count())
```

```python
new_x,new_y = dataset2.add_race_info(train_x,train_y,pcaed_df)
```
